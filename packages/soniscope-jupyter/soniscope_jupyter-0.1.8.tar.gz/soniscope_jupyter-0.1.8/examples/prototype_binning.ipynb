# 2D Histogram of Rectangular Bins 

```python
import numpy as np
import pandas as pd
```

```python
daily_file = 'bike_day_semantic.csv'
daily = pd.read_table(daily_file, sep=',')
daily
```

```python
df = daily
```

```python
stack = df.groupby([pd.cut(df.cnt, 3), pd.cut(df.windspeed, 3)]).size()
stack
```

```python
unstack = stack.unstack()
unstack
```

```python
stacksum = df.groupby([pd.cut(df.cnt, 3), pd.cut(df.windspeed, 3)]).casual.sum()
stacksum
```
